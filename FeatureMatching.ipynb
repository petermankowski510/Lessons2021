```python
"""Author: Peter Mankowski
ORB: An efficient alternative to SIFT or SURF"""

import numpy as np
import cv2
import matplotlib.pyplot as plt

"""Sample of one - quick hypothesis"""
img1 = cv2.imread('fry_set_train1.jpg',0)
img2 = cv2.imread('fry_set_test1.jpg',0)

orb = cv2.ORB_create()

kp1, des1 = orb.detectAndCompute(img1,None)
kp2, des2 = orb.detectAndCompute(img2,None)

bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
```

```python
matches = bf.match(des1,des2)
matches = sorted(matches, key = lambda x:x.distance)

img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches[:100],None, flags=2)

img3 = cv2.resize(img3, (1600,650)) 

plt.imshow(img3)
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Black spots only keypoints location,not size and orientation
img2 = cv2.drawKeypoints(img2, kp2, None, color=(0,255,0), flags=0)
plt.imshow(img2), plt.show()
```
