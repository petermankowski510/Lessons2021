```python
%matplotlib inline
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# creating two test data 
X = np.random.randint(10,35,(100,2)) 
Y = np.random.randint(55,70,(100,2)) 
Z = np.vstack((X,Y)) 
Z = Z.reshape((200,2)) 
  
# convert to np.float32 
Z = np.float32(Z) 
  
plt.xlabel('Test Data') 
plt.ylabel('Z samples') 
  
plt.hist(Z,256,[0, 256]) 
  
plt.show() 
```

```python
 # define criteria and apply kmeans() 
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0) 
ret,label,center = cv2.kmeans(Z,3,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS) 
  
# Now separate the data 
A = Z[label.ravel()==0] 
B = Z[label.ravel()==1] 
C = Z[label.ravel()==2] 
  
# Plot the data 
plt.scatter(A[:,0],A[:,1]) 
plt.scatter(B[:,0],B[:,1],c = 'r') 
plt.scatter(C[:,0],C[:,1],c = 'y') 

plt.scatter(center[:,0],center[:,1],s = 80,c = 'y', marker = 's') 
plt.xlabel('Test Data'),plt.ylabel('Z samples') 
plt.show() 
```
