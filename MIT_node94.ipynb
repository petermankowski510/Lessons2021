```python
"""
Digital microscope neuorlogical scan by team MIT34
Importing Image module from PIL package"""
import PIL
from PIL import Image  
import os, sys
import numpy as np
  
path1='C:/Users/seeho/datasets/aclImdb/cancer_ID09245.jpg'
im1 = Image.open(path1) 
im1 
```

```python
print(im1.format, im1.size, im1.mode)
```

```python
box = (100, 100, 240, 240)
region = im1.crop(box)
"""Crop and manimplate the R,G,B priot transposing steps
   Flip colors to identifiy the region later by another blocks"""
r, g, b = region.split()
region = Image.merge("RGB", (b, g, r))

region = region.transpose(Image.ROTATE_90)
im1.paste(region, box)
"""Save the altered frame here"""
im1.save("C:/Users/seeho/datasets/aclImdb/Alt_cancer_ID09245.jpg")

"""Show pixlel's R,G,B at that exact reference from the MIT98 microscope"""
r, g, b = im1.getpixel((110, 110))

print(" R,  G,  B: \n", r, g, b)
```

```python
"""display results side by side"""
image_orginal = Image.open('C:/Users/seeho/datasets/aclImdb/cancer_ID09245.jpg')
imge_augmented = Image.open('C:/Users/seeho/datasets/aclImdb/Alt_cancer_ID09245.jpg')
"""Temporary display to evaluate on the go"""
Image.fromarray(np.hstack((np.array(image_orginal),np.array(imge_augmented)))).show()
```
